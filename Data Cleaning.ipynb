```python
#ColNames = {1: "Serial Number", 2: "Pclass", 3: "Survived", 4: "Name", 5:"Gender", 6:"Age", 7:"Family", 8:"Fare", 9:"Embarked", 10:"Date"}
ColNames = list(["Serial Number", "Pclass", "Survived", "Name", "Gender", "Age", "Family", "Fare", "Embarked", "Date"])
df.columns=ColNames
```

```python
#Removing the first row of the dataset since we have already name the columns as per our requirement
df = df.drop(0)
```

```python
df.head(10)
```

```python
# We can see there are some missing values and special symbols in the age column. We need to fix these.

#Forcing the Age column to be string. So, we can manipulate it.
df['Age'] = df['Age'].astype(str) 

#Using Regular Expressions (regex) to strip out anything that isn’t a number or a decimal point.
df['Age'] = df['Age'].str.replace(r'[^\d\.]','', regex = True)

#Converting empty strings in Age column to true NaN values
df['Age'] = df['Age'].replace('', np.nan)

#Converting the column from string type tp foating type
df['Age'] = df['Age'].astype(float)
```

```python
df.head(10)
```

```python
#Cleaning the Gender column
# For row 4, the name contains "Mr.", so we can safely infer they are male!
df['Gender'] = df['Gender'].fillna(df['Name'].apply(lambda x: 'male' if 'Mr.' in str(x) or 'Master' in str(x) else 'female' if 'Mrs.' in str(x) or 'Miss' in str(x) else np.nan))
```

```python
### On the Titanic, a first-class passenger was generally older than a third-class immigrant. You can fill the missing ages based on the median age of their specific Passenger Class (Pclass) and Gender.
# Filling NaNs with median age of matching Pclass and Gender groups
df['Age'] = df['Age'].fillna(df.groupby(['Pclass', 'Gender'])['Age'].transform('median'))
```

```python
df.head(10)
```

```python
# Cleaning the Fare column
# We can see there are some special symbols (*) in the Fare column. We will replace them with true NaN values
import numpy as np
df['Fare'] = df['Fare'].replace(r'\*','',regex=True)
df['Fare'] = df['Fare'].replace('',np.nan,regex=True)
```

```python
df.head(10)
```

```python
# Filling NaN values with median Fare value grouped by Pclass and Gender
df['Fare'] = df['Fare'].fillna(df.groupby(['Pclass','Gender'])['Age'].transform('median'))
df.head(10)
```

```python
# Cleaning the Embarked column
# We can see that there are three values in the Embarked column - S, C, Q. However, there are some missing values. It is also possible that the values have spaces or are in lowercase. We need to fix these issues.

# Forcing the column to string type from object type so as to manipulate data as required
df['Embarked'] = df['Embarked'].astype(str)

# Stripping the string of left and right whitespaces
df['Embarked'] = df['Embarked'].str.strip()

# Converting all the values to uppercase
df['Embarked'] = df['Embarked'].str.upper()

# Replacing missing vlaues with true NaN values
df['Embarked'] = df['Embarked'].replace('',np.nan,regex=True)
```

```python
df.head(10)
```

```python
# Explicitly replacing the string 'NAN' with a true Python NaN object
df['Embarked'] = df['Embarked'].replace('NAN', np.nan)
# Replacing NaN values with the most common value (the mode of the Embarked column)
modeEmbarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(modeEmbarked)
df.head(10)
```

```python
# Cleaning the date column.
# We can see that there are "#######" strings at multiple places in the Date column.
# We will firstly replace them with true NaN values and then fill them with the most common date (mode of Date)
# We can also see that a format is being followed for the date, date-shortMonth- year, eg: 01-Jan-90

# Forcing the Date column to string type from onject type
df['Date'] = df['Date'].astype(str)

# Stripping the data of left and right whitespaces
df['Date'] = df['Date'].str.strip()

# Replacing # with empty strings
df['Date'] = df['Date'].replace(r'#', '', regex=False)

# Replacing empty values with true NaN values
df['Date'] = df['Date'].replace('', np.nan)

# Converting to date-time.
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y', errors = 'coerce')

# Handling the NaT values

# Finding the most common date (mode of Date)
mostCommonDate = df['Date'].mode()[0]

# Filling the missing dates with the most common date 
df['Date'] = df['Date'].fillna(mostCommonDate)
```
