### 1. Dropping irrelevant columns 
a. Serial Number  
b. Name  
c. Date

```python
import pandas as pd 
df = pd.read_csv('TheTitanicCleaned.csv')
df = df.drop(columns=['Serial Number', 'Name', 'Date'])

y = df['Survived']
```

```python
df.head(10)
```

### 2. Encoding categorical features into numbers
a. Gender   
b. Embarked 

```python
# Label encoding is good for binary data so we will use it for Gender encoding
# One Hot Encoding is good for categorical data with more than 2 categories
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
df['Gender'] = ohe.fit_transform(df[['Gender']])
df['Embarked'] = ohe.fit_transform(df[['Embarked']])
df.head(10)
```

### 3. Feature Scaling
a. Age  
b. Fare  
Columns like Age (ranging from 0 to 80) and Fare (ranging from 0 to 512) live on completely different scales. Algorithms like Logistic Regression or Support Vector Machines can get warped by large numbers.

We can use StandardScaler to normalize these features so they share a common variance scale.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Scale Age and Fare columns
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
df.head(10)
```

### Splitting into Training and Testing sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)
# 20% of the data will be used for training the model
print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
```
